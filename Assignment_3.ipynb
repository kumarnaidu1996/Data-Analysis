# 1.1 Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()


....Building my own reduce() function called my_reduce().

```python
def my_reduce(n):
    
    num_list = list(range(1,n+1)) #creating the list of numbers for a range of user inpu
    sum_number =0                      #initializing the sum_number with 0.
    for i in (num_list):               #using for loop to pic_up elements from num_list.
        sum_number+=i   
        
    return sum_number,num_list

#User input
print("Input:")
number = int(input("Enter the end number of a list you would like to Sum_up: "))

#function execution
kumar = my_reduce(number)

#output
print("List of the numbers are: ", kumar[1])
print("")
print("Output generated by my own reduce function: ")

print("Sum of all the elements in the list is: ",kumar[0])
```

....Generating the same output using built_in reduce() function.

```python
print("Input:")
number1 = int(input("Enter the number: "))
num_list1 = list(range(1,number1+1))
from functools import reduce              #import the reduce function

sum_number1 = reduce((lambda a,b:a+b),num_list1)

#output
print("List of the numbers are: ",num_list1)
print("")
print("Output by built_in redduce function: ")
print("Sum of all the elements in the list: ",sum_number1)
```

# 1.2 Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()

....My own filteer() function seperate even and odd elements of a list.

```python
#solutions_1.2
def my_filter(num):
    list_of_number = list(range(1,num+1))
    even_number=[]
    odd_number=[]
    for i in list_of_number:
        if i%2==0:
            even_number.append(i)
        else:
            odd_number.append(i)
    return list_of_number,even_number,odd_number

#input
print("Input:")
numbers = int(input("Enter the length of a list: "))

#funtion execution
split = my_filter(numbers)

#output
print("List of numbers: ",split[0])
print("")
print("Output generated by my_filter function")
print("List of even numbers: ",split[1])
print("List of add numbers: ",split[2])
```

....Printing the same output using the built_in filter() function.

```python
print("Input:")

numbers2 = int(input("Enter the length of list: "))

list_of_numbers2 = list(range(1,numbers2+1)) 

even = list(filter(lambda x:x%2==0 ,list_of_numbers2))
odd =  list(filter(lambda x:x%2!=0 , list_of_numbers2))

print("List of numbers: ",list_of_numbers2)
print("")

print("Output generated by built_in fiilter() function")
print("List of even number: ",even)
print("List of odd number: ",odd)
                        
```
